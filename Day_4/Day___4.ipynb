1-D prefix sum

```python
class Solution:
    def prefSum(self, arr):
        n = len(arr)
        prefix_sum = [0] * n
        prefix_sum[0] = arr[0]

        for i in range(1, n):
            prefix_sum[i] = prefix_sum[i - 1] + arr[i]
        return prefix_sum
```

## Mistakes i made
- Initialized prefix_sum with empty square braces [] like an empty list.
- multiplied with n, which does not make any sense

## Concept i understood
- prefix sum in 1D array is to add upto the specified index
- eg:

Input:
arr[] = [10, 20, 10, 5, 15]

Output: [10, 30, 40, 45, 60]

Explanation: For each index i, add all the elements from 0 to i:

prefixSum[0] = 10,

prefixSum[1] = 10 + 20 = 30,

prefixSum[2] = 10 + 20 + 10 = 40 and so on.

## Approach

- first initialise an auxillary array with n no:of 0's
- then initialise prefix_sum(auxillary arrar) as: prefix_sum[0] = arr[0]
- next run a loop from 1 to n
- inside the loop, add the current element with the previous element in the prefix_sum array and store it in the array, and the loop continues till end
- after the loop ends, return the array

## Kadane's algorithm

```python
class Solution:
    def maxSubarraySum(self, arr):
        n = len(arr)

        global_max = arr[0]
        current_max = arr[0]

        for i in range(1, n):
            current_max = max(arr[i], current_max + arr[i])

            global_max = max(current_max, global_max)

        return global_max
```

## Problem Faced

While implementing Kadane’s Algorithm, I initialized current_max and global_max to 0 and used
current_max = max(current_max, current_max + arr[i]).

## Issue

This logic never allowed starting a new subarray from the current element and also failed for arrays containing all negative numbers.

## Solution

I corrected the logic to compare arr[i] with current_max + arr[i] and initialized both variables with the first element of the array.

## Key Learning

Kadane’s Algorithm must always decide whether to start a new subarray or extend the existing one, and proper initialization is crucial for edge cases.

Max Sum Subarray of size K

```python
class Solution:
    def maxSubarraySum(self, arr, k):
        n = len(arr)

        window_sum = sum(arr[:k])

        max_sum = window_sum

        for i in range(n - k):
            window_sum = window_sum - arr[i] + arr[i + k]
            max_sum = max(max_sum, window_sum)

        return max_sum
```

## Problem

Find the maximum sum of a subarray of fixed size k using a contiguous window.

## Approach Used

Applied the Sliding Window Technique by:

Computing the sum of the first k elements

Sliding the window by removing the left element and adding the next element

## Solution

Maintained a running window_sum and updated max_sum at each step.

## Outcome

The solution works correctly and efficiently for all valid inputs.

## Key Learning

For fixed-size subarray problems, Sliding Window reduces time complexity from O(n·k) to O(n).
