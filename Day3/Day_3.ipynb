1. Linear Search


```python
class Solution:
    def search(self, arr, x):
        n = len(arr)
        for i in range(0, n):
            if arr[i] == x:
                return i
        return -1
```

2. Binary search


```python
class Solution:
    def binarysearch(self, arr, k):
        low = 0
        high = len(arr) - 1
        ans = -1
        while low <= high:

            mid  = low + (high - low) // 2

            if arr[mid] == k:
                ans = mid
                high = mid - 1

            if arr[mid] > k:
                high = mid - 1

            if arr[mid] < k:
                low = mid + 1
        return ans
```

3. First and Last Occurrences


```python
class Solution:
    def find(self, arr, x):

        first = -1
        last = -1
        n = len(arr)
        for i in range(0, n):
            if arr[i] == x:
                if first == -1:
                    first = i
                last = i
        return [first, last]
```

4. Number of occurrence

```python
class Solution:
    def countFreq(self, arr, target):
        n = len(arr)
        count = 0
        for i in range(0, n):
            if arr[i] == target:
                count += 1
        return count
```

5. Implement Upper Bound

```python
class Solution:
    def upperBound(self, arr, target):
        for i in range(0, len(arr)):
            if arr[i] > target:
                return i
            if i == len(arr) - 1:
                return len(arr)
```

6. Implement Lower Bound


```python
class Solution:
    def lowerBound(self, arr, target):
        for i in range(0, len(arr)):
            if arr[i] >= target:
                return i
        if i == len(arr) - 1:
            return len(arr)
```

7. Search in Rotated Sorted Array


```python
class Solution:
    def search(self, arr, key):
        for i in range(0, len(arr)):
            if arr[i] == key:
                return i
        return -1
```
