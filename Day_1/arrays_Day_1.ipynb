1. Reverse an Array

```python
class Solution:
    def reverseArray(self, arr):
        start = 0
        end = len(arr) - 1

        while start < end:
            arr[start], arr[end] = arr[end], arr[start]
            start += 1
            end -= 1

        return arr
```

2. Largest Element in Array

```python
class Solution:
    def largest(self, arr):
        max_num = arr[0]   # initialize from array to handle negatives

        for i in range(1, len(arr)):
            if arr[i] > max_num:
                max_num = arr[i]

        return max_num
```

3) Check if Array is Sorted

```python
class Solution:
    def isSorted(self, arr) -> bool:
        for i in range(len(arr) - 1):
            if arr[i] > arr[i + 1]:
                return False
        return True
```

4) Frequency of Elements

```python
class Solution:
    def countFreq(self, arr):
        arr.sort()
        result = []
        n = len(arr)
        i = 0

        while i < n:
            count = 1
            while i + 1 < n and arr[i] == arr[i + 1]:
                count += 1
                i += 1

            result.append([arr[i], count])
            i += 1

        return result
```
